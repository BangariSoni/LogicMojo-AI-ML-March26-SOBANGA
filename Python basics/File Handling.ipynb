```python
import csv

max_marks = 0
top_student = ""

with open("studentsRecord.csv", "r") as f:
    reader = csv.DictReader(f)
    
    for row in reader:
        marks = int(row["marks"])
        
        if marks > max_marks:
            max_marks = marks
            top_student = row["name"]

print("Top Student:", top_student)
print("Highest Marks:", max_marks)
```
