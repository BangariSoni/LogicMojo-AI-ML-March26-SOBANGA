```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def display(self):
        print("Name:", self.name)
        print("Marks:", self.marks)
s1 = Student("Soni", 95)
s1.display()
```

```python
class SimpleModel:
    def __init__(self):
        self.trained = False

    def fit(self):
        print("Training model...")
        self.trained = True

    def predict(self):
        if self.trained:
            print("Predicting...")
        else:
            print("Model not trained yet")
model = SimpleModel()
model.predict()   # Not trained
model.fit()
model.predict()   # Now works
```

```python
class Car:
    def __init__(self, color):
        self.color = color

    def drive(self):
        print("Car is driving")
c1 = Car("Red")
c1.drive()
```

```python
class Dog:
    def speak(self):
        print("Bark")

class Cat:
    def speak(self):
        print("Meow")
def make_sound(animal):
    animal.speak()

dog = Dog()
cat = Cat()
make_sound(dog)
make_sound(cat)
```
