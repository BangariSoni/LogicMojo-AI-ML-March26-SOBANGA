Mini Project: Student Marks Analyzer
Problem:

You have marks:
marks = [45, 78, 90, 34, 88, 76]

Task:

Print only passing marks (>= 40)

Skip failed marks

Calculate total of passing marks

Find average

Print grade:

Avg	Grade
≥ 90	A
≥ 70	B
≥ 50	C
else	D
 Hint:

Use:

for loop

continue

if-else

Expected Output (Example)
Passing Marks: 45 78 90 88 76
Average: 75.4
Grade: B


````python
marks = [45, 78, 90, 34, 88, 76]

passed = []
total = 0

for mark in marks:
    if mark >= 40:
        passed.append(mark)
        total += mark

avg = total / len(passed)

print("Passing Marks:", passed)
print("Average:", avg)

if avg >= 90:
    print("Grade: A")
elif avg >= 70:
    print("Grade: B")
elif avg >= 50:
    print("Grade: C")
else:
    print("Grade: D")
````

````python
#count failed students
marks = [45, 78, 90, 34, 88, 76]

passed = []
total = 0
count=0

for mark in marks:
    if mark >= 40:
        passed.append(mark)
        total += mark
    else:
        count+=1


if len(passed) > 0:
    avg = total / len(passed)
else:
    avg = 0

print("Passing Marks:", passed)
print("Failed Students",count)
print("Average:", avg)
passed.sort()
print("Sorted Marks:",passed)
print("Highest Mark:", passed[-1])
print("Lowest Mark:", passed[0])

if avg >= 90:
    print("Grade: A")
elif avg >= 70:
    print("Grade: B")
elif avg >= 50:
    print("Grade: C")
else:
    print("Grade: D")
````

````python
#approach 1
nums = [5, 3, 8, 1, 9, 2, 7]

first = max(nums)
second = float('-inf')

for num in nums:
    if num != first and num > second:
        second = num

print(second)
#approach 2
nums = [5, 3, 8, 1, 9, 2, 7]
first = second = float('-inf')
for num in nums:
    if num > first:
        second = first
        first = num
    elif num > second and num != first:
        second = num
print(second)
````

````python
# Trick 1 — setdefault() → add key only if not exists
d = {"name": "Alice"}
d.setdefault("age", 25)     # adds age=25 ✅
d.setdefault("name", "Bob") # name already exists → ignored!
print(d)  # {"name": "Alice", "age": 25}

# Trick 2 — merge two dicts ⭐
d1 = {"a": 1, "b": 2}
d2 = {"c": 3, "d": 4}

# Python 3.9+ way
merged = d1 | d2    # {"a":1, "b":2, "c":3, "d":4}

# Older way
merged = {**d1, **d2}   # same result ✅
print(merged)
# Trick 3 — count frequency of items ⭐
words = ["apple", "banana", "apple", "mango", "banana", "apple"]
freq = {}
for word in words:
    freq[word] = freq.get(word, 0) + 1
print(freq)  # {"apple": 3, "banana": 2, "mango": 1}

# Trick 4 — group items ⭐
students = [("Alice", "A"), ("Bob", "B"), ("Charlie", "A"), ("Diana", "B")]
groups = {}
for name, grade in students:
    groups.setdefault(grade, []).append(name)
print(groups)  # {"A": ["Alice", "Charlie"], "B": ["Bob", "Diana"]}
````

````python
# Q3 — Medium
# pythonwords = ["apple", "banana", "apple", "mango", "banana", "apple", "mango", "mango"]
# ```
# **Count frequency of each word using a dictionary and print them sorted by frequency in descending order.**

# Expected output:
# ```
# apple  → 3
# mango  → 3
# banana → 2
pythonwords = ["apple", "banana", "apple", "mango", "banana", "apple", "mango", "mango"]
freq={}
for word in pythonwords:
    freq[word]=freq.get(word,0)+1
print(sorted(freq,key=freq.get, reverse=True))
#OR_____________
for word in sorted(freq, key=freq.get, reverse=True):
    print(f"{word} → {freq[word]}")

# apple  → 3
# mango  → 3
# banana → 2
# If you want alphabetical order for ties:
sorted(freq, key=lambda x: (-freq[x], x), reverse=False)
# (-3, "apple")  → apple first  ✅
# (-3, "mango")  → mango second ✅
# (-2, "banana") → banana last  ✅
````

````python
data = [
    (101, "Rahul", 80000, "India", "Fraud"),
    (102, "Amit", 30000, "India", "Safe"),
    (103, "John", 90000, "USA", "Fraud"),
    (101, "Rahul", 80000, "India", "Fraud"),
    (104, "Sara", 40000, "USA", "Safe"),
    (105, "Mike", 120000, "USA", "Fraud")
]

# 1. Remove duplicates
unique_data = list(set(data))

# 2. Convert to dictionary format
data_dict = {}
for record in unique_data:
    data_dict.setdefault(record[0], []).append(record[1:])

# 3-8. Process customers
customers = []
countries = set()
fraud_count = 0
results = []

for row in unique_data:
    id, name, income, country, status = row

    # 6. Skip low income
    if income < 40000:
        continue

    # 7. Stop if > 1 lakh ← debatable, continue is safer!
    if income > 100000:
        print(f"Very high value customer found: {name}")
        break

    # 2. Dictionary format for row
    customer = {"id": id, "name": name, "country": country, "status": status}
    customers.append(customer)

    # 3. Unique countries
    countries.add(country)

    # 5. Count fraud
    if status == "Fraud":
        fraud_count += 1

    # 4. High income label
    if income > 80000:
        label = "High risk and high value" if status == "Fraud" else "High value safe"
    else:
        label = "Normal"

    # 8. Store as tuple
    results.append((id, label))
print("Dictionary",data_dict)
print("Customers(each customer):", customers)
print("Countries:", countries)
print("Fraud count:", fraud_count)
print("Results:", results)
print("Data dict:", data_dict)
````
