```python
name = "banana"
name = list(name)
print(name)

for i in range(len(name)):
    if name[i] == "*":
        continue

    count = 1

    for j in range(i + 1, len(name)):
        if name[i] == name[j]:
            count += 1
            name[j] = "*"   # mark visited

    if count > 1:
        print(name[i], "->", count)
```

```python
name = "banbana"
visited = ""

for i in range(len(name)):
    if name[i] in visited:
        continue

    count = 1

    for j in range(i + 1, len(name)):
        if name[i] == name[j]:
            count += 1
            print("vis",visited,count)

    visited += name[i]
    print("outer vis",visited,count)

    if count > 1:
        print(name[i], "->", count)
```

Mini Project: Student Marks Analyzer
Problem:

You have marks:
marks = [45, 78, 90, 34, 88, 76]

Task:

Print only passing marks (>= 40)

Skip failed marks

Calculate total of passing marks

Find average

Print grade:

Avg	Grade
≥ 90	A
≥ 70	B
≥ 50	C
else	D
 Hint:

Use:

for loop

continue

if-else

Expected Output (Example)
Passing Marks: 45 78 90 88 76
Average: 75.4
Grade: B


```python
marks = [45, 78, 90, 34, 88, 76]

passed = []
total = 0

for mark in marks:
    if mark >= 40:
        passed.append(mark)
        total += mark

avg = total / len(passed)

print("Passing Marks:", passed)
print("Average:", avg)

if avg >= 90:
    print("Grade: A")
elif avg >= 70:
    print("Grade: B")
elif avg >= 50:
    print("Grade: C")
else:
    print("Grade: D")
```

```python
#count failed students
marks = [45, 78, 90, 34, 88, 76]

passed = []
total = 0
count=0

for mark in marks:
    if mark >= 40:
        passed.append(mark)
        total += mark
    else:
        count+=1


if len(passed) > 0:
    avg = total / len(passed)
else:
    avg = 0

print("Passing Marks:", passed)
print("Failed Students",count)
print("Average:", avg)
passed.sort()
print("Sorted Marks:",passed)
print("Highest Mark:", passed[-1])
print("Lowest Mark:", passed[0])

if avg >= 90:
    print("Grade: A")
elif avg >= 70:
    print("Grade: B")
elif avg >= 50:
    print("Grade: C")
else:
    print("Grade: D")
```

```python
transactions = [1200, 3000, 250, 7800, 540, 3000, 9100, 250, 1500, 7800]
visited=[]
more_than_once=[]
fraud_detection=[]
total=0
for i in range(len(transactions)):
    total+=transactions[i]
    if transactions[i]>5000:
        fraud_detection.append(transactions[i])
    if transactions[i] in visited:
        continue
    count =1
    for j in range(i+1,len(transactions)):
        if transactions[i]==transactions[j]:
            count+=1
    visited.append(transactions[i])
    if count>1:
        more_than_once.append(transactions[i])
print("Transaction values that appear more than once:",more_than_once)
print("Fraud Monitoring transaction:",fraud_detection)
print("Total spending of the day:",total)
print("Unique Transaction",visited)
visited.sort(reverse=True)
print("Unique Transaction sorted in desc",visited)

```
