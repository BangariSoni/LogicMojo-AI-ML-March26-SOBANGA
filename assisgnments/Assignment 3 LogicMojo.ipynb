    Q1. Understanding Feature Distributions (Iris Dataset)

    Context: Before training any ML model, we analyze feature distributions.

    Dataset: Load Iris dataset from sklearn

    Tasks:

    1. Extract feature matrix as NumPy array
    2. Compute:
    Mean
    Median
    Standard deviation
    Variance (for each feature)
    3. Identify:
    Which feature has highest variability and why it matters in ML
    4. Convert any one feature into shape (n,1) and explain why ML models expect this format

```python
from sklearn.datasets import load_iris
import numpy as np

# Load dataset
iris = load_iris()

# Task 1 — Extract feature matrix as NumPy array
X = iris.data
print("Shape of X:", X.shape)
print("Feature Names:", iris.feature_names)

# Task 2 — Compute statistics for each feature
mean = np.mean(X, axis=0)
median = np.median(X, axis=0)
std = np.std(X, axis=0)
variance = np.var(X, axis=0)

print("\n--- Statistics per Feature ---")
for i, name in enumerate(iris.feature_names):
    print(f"\n{name}:")
    print(f"  Mean     : {mean[i]:.4f}")
    print(f"  Median   : {median[i]:.4f}")
    print(f"  Std      : {std[i]:.4f}")
    print(f"  Variance : {variance[i]:.4f}")

# Task 3 — Feature with highest variability
highest_var_index = np.argmax(variance)
highest_var_feature = iris.feature_names[highest_var_index]

print("\n--- Highest Variability ---")
print(f"Feature with highest variance : {highest_var_feature}")
print(f"Variance value                : {variance[highest_var_index]:.4f}")
print("Explanation: petal_length varies greatly across species.This makes it the most useful feature for classification.")
# Task 4 — Reshape one feature to (n, 1)
petal_length = X[:, 2]  # extract 3rd column (petal_length)
print("\nBefore reshape:", petal_length.shape)

petal_length_2d = petal_length.reshape(-1, 1)
print("After reshape :", petal_length_2d.shape)
print("Why (n,1) format explanation: ML models expect 2D input (samples x features).")
print("Shape (150,) is 1D — ambiguous. Shape (150,1) is unambiguous.")
```

    Q11. End-to-End ML Preprocessing Pipeline (Titanic)

    Context: Prepare dataset for ML model training

    Tasks:

    1. Load dataset
    2. Handle missing values
    3. Encode categorical variables (sex, embarked)
    4. Normalize numerical features
    5. Convert to NumPy array

    Output:

    Final feature matrix ready for model training

```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import StandardScaler

# 1️Load Dataset
df = sns.load_dataset("titanic")

print("Original Shape:", df.shape)
print("\nMissing Values:\n", df.isnull().sum())


# --------------------------------------------------
# 2️Handle Missing Values
# --------------------------------------------------

# Fill Age with median (robust to outliers)
df["age"] = df["age"].fillna(df["age"].median())

# Fill Embarked with most frequent value (mode)
df["embarked"] = df["embarked"].fillna(df["embarked"].mode()[0])

# Drop Cabin (too many missing values)
df = df.drop(columns=["deck"])

print("\nMissing Values After Handling:\n", df.isnull().sum())


# --------------------------------------------------
# 3️Encode Categorical Variables
# --------------------------------------------------

# Encode 'sex' (Label Encoding style)
df["sex"] = df["sex"].map({"male": 0, "female": 1})

# One-hot encode 'embarked'
df = pd.get_dummies(df, columns=["embarked"], drop_first=True)

print("\nColumns After Encoding:\n", df.columns)


# --------------------------------------------------
# 4️Select Important Features
# --------------------------------------------------

features = ["pclass", "sex", "age", "sibsp", "parch", "fare",
            "embarked_Q", "embarked_S"]

X = df[features]
y = df["survived"]


# --------------------------------------------------
# 5️Normalize Numerical Features
# --------------------------------------------------

scaler = StandardScaler()

numerical_cols = ["age", "fare"]

X[numerical_cols] = scaler.fit_transform(X[numerical_cols])

print("\nMean After Scaling:\n", X[numerical_cols].mean())
print("\nStd After Scaling:\n", X[numerical_cols].std())


# --------------------------------------------------
# 6️Convert to NumPy Array
# --------------------------------------------------

X_final = X.values
y_final = y.values

print("\nFinal Feature Matrix Shape:", X_final.shape)
print("Final Target Shape:", y_final.shape)
```

Q12. Feature Selection using Statistical Understanding

Context: Not all features improve model performance.

Dataset: Iris

Tasks:

1. Compute correlation matrix
2. Identify highly correlated features
3. Drop redundant features
4. Explain:
Impact on overfitting and model performance

```python
import pandas as pd
import seaborn as sns
import numpy as np

df = sns.load_dataset("iris")

X = df.drop("species", axis=1)

corr_matrix = X.corr()

high_corr_pairs = []

threshold = 0.9

for i in range(len(corr_matrix.columns)):
    for j in range(i):
        if abs(corr_matrix.iloc[i, j]) > threshold:
            high_corr_pairs.append((corr_matrix.columns[i], corr_matrix.columns[j]))

cols_to_drop = set([pair[0] for pair in high_corr_pairs])

X_reduced = X.drop(columns=list(cols_to_drop))

print("Correlation Matrix:\n", corr_matrix)
print("\nHighly Correlated Feature Pairs:", high_corr_pairs)
print("\nDropped Columns:", cols_to_drop)
print("\nReduced Feature Set Shape:", X_reduced.shape)
```

Q13. Custom Feature Transformation using apply()

Context: Real ML requires custom logic transformation.

Dataset: Titanic

Tasks:

1. Create age_group:
Child / Adult / Senior
2. Use .apply()
3. Compute survival rate per group
4. Interpret:
Which segment has highest survival likelihood

```python
import pandas as pd
import seaborn as sns
import numpy as np

df = sns.load_dataset("titanic")

df["age"] = df["age"].fillna(df["age"].median())

def categorize_age(age):
    if age < 18:
        return "Child"
    elif age < 60:
        return "Adult"
    else:
        return "Senior"

df["age_group"] = df["age"].apply(categorize_age)

survival_by_group = df.groupby("age_group")["survived"].mean()

print("Survival Rate by Age Group:\n", survival_by_group)
print("\nHighest Survival Group:", survival_by_group.idxmax())
```

Q14. Error Analysis Simulation (Model Evaluation Thinking)

Context: Understanding prediction errors is key in ML.

Tasks:

1. Generate 1000 random error values using NumPy
2. Compute:
Mean
Standard deviation
3. Identify outliers:
Using mean ± 2*std
4. Explain:
What high variance means in model performance

```python
import numpy as np

np.random.seed(42)

errors = np.random.normal(loc=0, scale=1, size=1000)

mean_error = np.mean(errors)
std_error = np.std(errors)

lower_bound = mean_error - 2 * std_error
upper_bound = mean_error + 2 * std_error

outliers = errors[
    (errors < lower_bound) | (errors > upper_bound)
]

print("Mean Error:", mean_error)
print("Standard Deviation:", std_error)
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
print("Number of Outliers:", len(outliers))
```
