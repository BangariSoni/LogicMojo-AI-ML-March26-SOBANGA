Q3. Feature Scaling using NumPy (Very Important)

Context: Most ML models require normalized data.

Dataset: Iris (NumPy array)

Tasks:

1. Apply standardization:

$$
X' = \frac{X - \mu}{\sigma}
$$
	​2. 

Verify:

```python
import numpy as np
# Extract feature matrix as NumPy array
X = df.drop("species", axis=1).to_numpy()
# Compute column-wise mean and std
mu = np.mean(X, axis=0)
sigma = np.std(X, axis=0)
# Apply Standardization
X_scaled = (X - mu) / sigma
# Verify results
mean_after = np.mean(X_scaled, axis=0)
std_after = np.std(X_scaled, axis=0)
print("Original Shape:", X.shape)
print("Mean after scaling:", mean_after)
print("Std after scaling:", std_after)
```

Q8. Group-Based Insights (Customer Segmentation Logic)

Context: ML models often depend on aggregated patterns.

Average tip by day
Total bill by gender
2. Create new feature:

3. Identify:

4. Explain:

```python
import pandas as pd
import seaborn as sns

df = sns.load_dataset("tips")

avg_tip_by_day = df.groupby("day")["tip"].mean()

total_bill_by_gender = df.groupby("sex")["total_bill"].sum()

df["tip_percentage"] = df["tip"] / df["total_bill"] * 100

highest_tipping_day = df.groupby("day")["tip_percentage"].mean().idxmax()

print("Average Tip by Day:\n", avg_tip_by_day)
print("\nTotal Bill by Gender:\n", total_bill_by_gender)
print("\nHighest Tipping Day:", highest_tipping_day)
```

```python
import numpy as np
A = np.array([[1, 2],[3, 4],[5, 6]])
B = np.array([[7, 8],[9, 10],[11, 12]])

vertical_stack = np.vstack((A, B))

horizontal_stack = np.hstack((A, B))

reshaped_array = horizontal_stack.reshape(2, 6)

print("A shape:", A.shape)
print("B shape:", B.shape)
print("Vertical Stack shape:", vertical_stack.shape)
print("Horizontal Stack shape:", horizontal_stack.shape)
print("Reshaped (2,6) shape:", reshaped_array.shape)
```

    2. Handle missing values

    4. Normalize numerical features

Q12. Feature Selection using Statistical Understanding

Context: Not all features improve model performance.

Dataset: Iris

Tasks:

1. Compute correlation matrix
2. Identify highly correlated features
3. Drop redundant features
4. Explain:
Impact on overfitting and model performance

```python
import pandas as pd
import seaborn as sns
import numpy as np

df = sns.load_dataset("iris")

X = df.drop("species", axis=1)

corr_matrix = X.corr()

high_corr_pairs = []

threshold = 0.9

for i in range(len(corr_matrix.columns)):
    for j in range(i):
        if abs(corr_matrix.iloc[i, j]) > threshold:
            high_corr_pairs.append((corr_matrix.columns[i], corr_matrix.columns[j]))

cols_to_drop = set([pair[0] for pair in high_corr_pairs])

X_reduced = X.drop(columns=list(cols_to_drop))

print("Correlation Matrix:\n", corr_matrix)
print("\nHighly Correlated Feature Pairs:", high_corr_pairs)
print("\nDropped Columns:", cols_to_drop)
print("\nReduced Feature Set Shape:", X_reduced.shape)
```

Q13. Custom Feature Transformation using apply()

Context: Real ML requires custom logic transformation.

Dataset: Titanic

Tasks:

1. Create age_group:
Child / Adult / Senior
2. Use .apply()
3. Compute survival rate per group
4. Interpret:
Which segment has highest survival likelihood

```python
import pandas as pd
import seaborn as sns
import numpy as np

df = sns.load_dataset("titanic")

df["age"] = df["age"].fillna(df["age"].median())

def categorize_age(age):
    if age < 18:
        return "Child"
    elif age < 60:
        return "Adult"
    else:
        return "Senior"

df["age_group"] = df["age"].apply(categorize_age)

survival_by_group = df.groupby("age_group")["survived"].mean()

print("Survival Rate by Age Group:\n", survival_by_group)
print("\nHighest Survival Group:", survival_by_group.idxmax())
```

Q14. Error Analysis Simulation (Model Evaluation Thinking)

Context: Understanding prediction errors is key in ML.

Tasks:

1. Generate 1000 random error values using NumPy
2. Compute:
Mean
Standard deviation
3. Identify outliers:
Using mean ± 2*std
4. Explain:
What high variance means in model performance

```python
import numpy as np

np.random.seed(42)

errors = np.random.normal(loc=0, scale=1, size=1000)

mean_error = np.mean(errors)
std_error = np.std(errors)

lower_bound = mean_error - 2 * std_error
upper_bound = mean_error + 2 * std_error

outliers = errors[(errors < lower_bound) | (errors > upper_bound)]

print("Mean Error:", mean_error)
print("Standard Deviation:", std_error)
print("Number of Outliers:", len(outliers))
```
